```python
# Import semua modul yang dibutuhkan
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
# Buat DataFrame dari dataset csv
df = pd.read_csv('https://raw.githubusercontent.com/mfadlirahmat/Fadli-s-Dataset/main/Pima%20Indians%20diabetes.csv')
```

```python
df.head()
```

```python
# cek apakah ada missing values
df.info()
```

```python
# Deklarasi atribut dan label
x = df[df.columns[:8]]
y = df['Outcome']
```

```python
# Melakukan standardisasi atribut
scaler = StandardScaler()
scaler.fit(x)
x= scaler.transform(x)
```

```python
# Membagi data menjadi data train dan test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
# Melakukan pelatihan training dengan model SVC
clf = SVC()
clf.fit(x_train, y_train)
```

```python
# Mengecek akurasi model
clf.score(x_test, y_test)
```

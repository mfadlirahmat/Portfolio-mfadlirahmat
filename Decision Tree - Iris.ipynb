```python
#import library yang dibutuhkan
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
```

```python
# Membuat Dataframe dari file csv
df = pd.read_csv('https://raw.githubusercontent.com/mfadlirahmat/Fadli-s-Dataset/main/Iris.csv')
```

```python
df.head()
```

```python
df.drop('Id', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Deklarasi Atribut dan Label
x = df[['SepalLengthCm', 'SepalWidthCm',	'PetalLengthCm',	'PetalWidthCm']]
y = df['Species']
```

```python
# Melakukan pelatihan pada model
tree_model = DecisionTreeClassifier()
tree_model.fit(x,y)
```

```python
# Mencoba melakukan prediksi
tree_model.predict([[6.2, 3.4, 5.4, 2.3]])
```

```python
# membuat visualisasi decision tree
export_graphviz(
    tree_model,
    out_file ='iris_tree.dot',
    feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
    class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
    rounded = True,
    filled = True
)
```

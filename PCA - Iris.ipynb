```python
# import modul-modul yang dibutuhkan
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.decomposition import PCA
```

```python
# load dataset iris
iris = datasets.load_iris()
```

```python
# mendeklarasi atribut dan label
atribut = iris.data
label = iris.target
```

```python
# membagi data training dan test
x_train, x_test, y_train, y_test = train_test_split(atribut, label, test_size =0.2, random_state=1)
```

```python
# Coba langsung ditraining dengan metode Decision Tree
decision_tree = tree.DecisionTreeClassifier()
model_pertama = decision_tree.fit(x_train, y_train)
model_pertama.score(x_test, y_test)
```

```python
# Persiapan PCA Analysis
pca = PCA(n_components=4)
pca_atributes = pca.fit_transform(x_train)
```

```python
# Menampilkan hasil Analysis PCA
pca.explained_variance_ratio_
```

```python
# Pake dua atribut sudah cukup significant
# Transform semua atribut
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.fit_transform(x_test)
```

```python
# Training atribut yang sudah ditransform dengan Decision Tree
model2 = decision_tree.fit(x_train_pca, y_train)
model2.score(x_test_pca, y_test)
```
